```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib 
import matplotlib.pyplot as plt 
plt.style.use("ggplot")
from matplotlib import figure 
%matplotlib inline
matplotlib.rcParams['figure.figsize']=(14,8)

```

```python
data=pd.read_csv(r'C:\Users\ibrahim\Desktop\GDP_new.csv')
```

```python
data["country"]=data["country"].astype("string")
data["year"]=data["year"].astype("int32")
```

```python
data.dtypes
```

```python
sns.scatterplot(data=data, x="life_expectancy", y="gdp_pp") 
plt.title("Life Expectancy vs Gdp Per Person")
plt.show()
```

```python
sns.barplot(x="life_expectancy", y="gdp_pp", data=data)
plt.title("Life Expectancy vs Gdp Per Person")
plt.xticks(rotation="vertical")
plt.show()
```

```python
sns.regplot(data=data,x="life_expectancy",y="gdp_pp",scatter_kws={"color":"red"},line_kws={"color":"blue"})
plt.title("Life Expectancy vs Gdp Per Person")
plt.show()
```

```python
correlation=data.corr(method="pearson") # I choose pearson method out of (kendall,pearson,spearman) methods
correlation
```

```python
sns.heatmap(correlation,annot=True)
plt.title("Correlation between Life Expectancy vs Gdp")
plt.show()
```

```python
unstacked_corr=correlation.unstack()
unstacked_corr
```

```python
unstacked_corr[unstacked_corr > 0.5 ] # As we see there is high correlation between Gdp and Life Expectancy according to 'pearson' method
```
